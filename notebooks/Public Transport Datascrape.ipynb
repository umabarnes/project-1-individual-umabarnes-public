```python
import pandas as pd
import requests
```

```python
df = pd.read_csv("../data/daily_ridership.csv")
```

```python
df.head()
```

```python
print(f"Number of instances: {len(df)}")
print(f"Number of features : {len(df.columns)}")
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df = df[(df['Date'] >= '2023/07/01') & (df['Date'] < '2024/01/01')]
```

```python
print(f"Number of instances within required period: {len(df)}")
```

```python
df.head()
```

```python
df.tail()
```

```python
# remove irrelevant columns which compare to pre-pandemic day
columns_to_drop = [col for col in df.columns if '% of Comparable Pre-Pandemic Day' in col]

df = df.drop(columns=columns_to_drop)
df = df.drop('LIRR: Total Estimated Ridership', axis=1)

len(df)
```
