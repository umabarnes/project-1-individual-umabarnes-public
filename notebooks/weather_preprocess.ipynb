## Weather Data

```python
import pandas as pd
import requests
import json
import numpy as np
import datetime as dt
```

```python
df = pd.read_csv('https://www.ncei.noaa.gov/data/global-hourly/access/2023/72505394728.csv')
df.tail()
```

- this warning is irrelevant for our analysis as we will be removing columns (36,42,43,48,52,88) anyway

```python
print(f"Number of instances: {len(df)}")
print(f"Number of features : {len(df.columns)}")
```

```python
# Narrow down to the required period
df = df[(df['DATE'] >= '2023-07-01') & (df['DATE'] < '2024-01-01')]
df
```

```python
print(f"Number of instances within required period: {len(df)}")
```

```python
df_copy = df.copy()
df_copy = df.loc[df['REPORT_TYPE'] == 'FM-15', :]

# Extract the unscaled values for each column
df_copy.loc[:,'WND'] = (df_copy['WND'].apply(lambda x: int(x.split(',')[-2])/10)
                            .replace(999.9, np.nan))
    
df_copy.loc[:,'TMP'] = (df_copy['TMP'].apply(lambda x: int(x.split(',')[0])/10)
                            .replace(999.9, np.nan))
    
df_copy.loc[:,'DEW'] = (df_copy['DEW'].apply(lambda x: int(x.split(',')[0])/10)
                            .replace(999.9, np.nan))
                        
```

```python
# Impute missing data using data from an hour after
df_copy.bfill(inplace=True)
```

```python
# Extract date and hour from datetime column
df_copy.loc[:,'date'] = pd.to_datetime(df_copy['DATE'])
df_copy.loc[:,'hour'] = df_copy['date'].dt.hour
df_copy.loc[:,'date'] = df_copy['date'].dt.date

# Rename columns and put in lowercase for consistency
df_copy.rename({'WND':'wnd',
                        'TMP':'tmp',
                        'DEW':'dew'},
                        axis=1,
                        inplace=True)

df_copy = df_copy[['date', 'hour', 'wnd', 'tmp', 'dew']]
df_copy = df_copy.reset_index(drop=True)
```

```python
df_copy
```

```python

# Convert 'date' column to datetime
df_copy['date'] = pd.to_datetime(df_copy['date'])

# Remove duplicates
df_copy = df_copy.drop_duplicates(subset=['date', 'hour'])

# Create the full range of dates and hours
date_range = pd.date_range(start='2023-07-01', end='2023-12-31', freq='D')
hour_range = range(0, 24)
full_index = pd.MultiIndex.from_product([date_range, hour_range], names=['date', 'hour'])

# Reindex the original DataFrame to include all date-hour combinations
df_complete = df_copy.set_index(['date', 'hour']).reindex(full_index).reset_index()

# Find missing rows
missing_rows = df_complete[df_complete.isna().any(axis=1)]

# Display missing date-hour combinations
print(missing_rows)
```

```python
df_complete_filled = df_complete.interpolate(method='linear')
df_complete_filled
```

```python
nan_counts = df_complete_filled.isna().sum()
nan_counts
```

```python
# saving hourly weather to directory
df_complete_filled.to_csv('../data/curated/hourly_weather.csv', index=False)
```

```python

# Group by date and calculate daily averages
daily_weather_df = df_complete_filled.groupby('date').agg({
    'wnd': 'mean',   # Average wind speed
    'tmp': 'mean',   # Average temperature
    'dew': 'mean',   # Average dew point
}).reset_index()

len(daily_weather_df)
```
